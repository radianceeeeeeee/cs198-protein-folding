```python
from ipynb.fs.full.dijkstra import shortest_path

import ipynb.fs.full.weight_functions as weights 
import ipynb.fs.full.matrix_generator as matrix_gen
import numpy as np
```

```python
m = matrix_gen.generate_matrix()
```

```python
x = (2, 24)
y = (15, 1)

print(f'{x}: {m[x[0]][x[1]]}')
print(f'{y}: {m[y[0]][y[1]]}')

dist, path = shortest_path(m, x, y, weights.euclidean_distance_3d)

print(f'Total distance: {dist}')
print("Path: ", end="")
print(*path, sep=" -> ")
```

```python
matrix_gen.generate_2d_map_with_path(m, path)
```
